### Create Dataframe

```python
employee_Data = [(10, "Raj", "Kumar", "1999", "100", "M", 2000),
                 (20, "Rahul", "Ranjan", "2002", "200", "F", 8000),
                 (30, "Raghav", "Manish", "2010", "100", None, 6000),
                 (40, "Raja", "Singh", "2004", "100", "F", 7000),
                 (50, "Rama", "Krish", "2008", "400", "M", 1000),
                 (60, "Rasul", "Kutty", "2014", "500", "M", 5000),
                 (70, "Kumar", "Chand", "2004", "600", "M", 5000),
                  (80, "Kumari", "Krishna", "2007", "800", "F", 4000)
                 ]

employee_Schema = ["Employee_ID", "First_Name", "Last_Name", "DOJ", "Employee_Department_ID","Gender", "Salary"]

employee_DF = spark.createDataFrame(data = employee_Data, schema = employee_Schema)
display(employee_DF)
```

### Create Department Dataframe

```python
department_Data = [("HR", 100),
                   ("Supply", 200),
                   ("Sales", 300),
                   ("Stock", 400),
                   ("Finance", 500),
                   ("Marketing", 600),
                   ("IT", 700),
                   ]

department_Schema = ["Department", "Department_ID"]

department_DF = spark.createDataFrame(data = department_Data, schema = department_Schema)
display(department_DF)
```

### Inner Join

```python
df_InnerJoin = employee_DF.join(department_DF, employee_DF.Employee_Department_ID == department_DF.Department_ID, "inner")
display(df_InnerJoin)
```

### Full Outer Join

```python
df_fullOuterJoin = employee_DF.join(department_DF, employee_DF.Employee_Department_ID == department_DF.Department_ID, "full_outer")
display(df_fullOuterJoin)
```

### Left Outer Join

```python
df_leftOuterJoin = employee_DF.join(department_DF, employee_DF.Employee_Department_ID == department_DF.Department_ID, "left_outer")
display(df_leftOuterJoin)
```

### Right Outer Join


```python
df_rightOuterJoin = employee_DF.join(department_DF, employee_DF.Employee_Department_ID == department_DF.Department_ID, "right_outer")
display(df_rightOuterJoin)
```

### Left Semi Join

```python
df_leftSemiJoin = employee_DF.join(department_DF, employee_DF.Employee_Department_ID == department_DF.Department_ID, "left_semi")
display(df_leftSemiJoin)
```

### Left Anti Join

```python
df_leftAntiJoin = employee_DF.join(department_DF, employee_DF.Employee_Department_ID == department_DF.Department_ID, "left_anti")
display(df_leftAntiJoin)
```
