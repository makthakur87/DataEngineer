```python
from delta.tables import *

DeltaTable.create(spark) \
  .tableName('person_demo') \
  .addColumn('emp_id', 'INT') \
  .addColumn('emp_name', 'STRING') \
  .addColumn('gender', 'STRING') \
  .addColumn('salary', 'INT') \
  .addColumn('dept', 'STRING') \
  .property('description', 'table created for demo purpose') \
  .execute()
```

```python
%sql
insert into person_demo values(100, 'Stephen', 'M', 2000, 'IT');
insert into person_demo values(200, 'Phillip', 'M', 8000, 'HR');
insert into person_demo values(300, 'Lara', 'F', 6000, 'SALES');
```

```python
%sql
select * from person_demo
```

```python
deltaInsance = DeltaTable.forPath(spark, 'workspace.default.person_demo')
```

```python
display(deltaInsance.toDF())
```

```python
%sql
delete  from person_demo where emp_id = 100
```

```python
%sql
select * from workspace.default.person_demo
```

```python
deltaInsance.delete('emp_id = 200')
```
