### Create Sample Code

```python
csvFile = "/Volumes/workspace/default/mandy/baby_names/Baby_Names_2007_2009.csv"

csvDF = (spark.read
         .option('sep', ",")
         .option("inferSchema", "true")
         .option("header", "true")
         .csv(csvFile))

display(csvDF)
```

### Snappy Compression

```python
csvDF.write.format("parquet").option("compression", "snappy").save("/Volumes/workspace/default/mandy/stream_write/snappy.parquet")
```

### Gzip Compression

```python
csvDF.write.format("parquet").option("compression", "gzip").save("/Volumes/workspace/default/mandy/stream_write/gzip.parquet")
```
