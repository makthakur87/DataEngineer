### Create First Dataframe

```python
empData1 = [
  (111, 'Stephen', 'King', 2000),
  (222, 'Philip', 'Larkin', 8000),
  (333, 'John', 'Smith', 6000)
]

empSchema1 = ['Id', 'Firstname', 'Lastname', 'Salary']

df1 = spark.createDataFrame(empData1, empSchema1)
display(df1)
```

### create Second Dataframe with Same Schema as FirstDataframe

```python
empData2 = [
  (444, 'Thomas', 'Frank', 4000),
  (555, 'Stephen', 'Fleming', 3000),
  (666, 'William', 'Pending', 7000)
]

empSchema2 = ['Id', 'Firstname', 'Lastname', 'Salary']

df2 = spark.createDataFrame(empData2, empSchema2)
display(df2)
```

### Create Third Dataframe with Different Schema from First Dataframe

```python
empData3 = [
  (777, 'David', 4000),
  (888, 'Mike', 3000),
  (999, 'Winsten', 7000)
]

empSchema3 = ['Id', 'Name', 'Salary']

df3 = spark.createDataFrame(empData3, empSchema3)
display(df3)
```

### Compare Schema of First and Second Dataframe

```python
if df1.schema == df2.schema:
    print('schema matches')
else:
    print('schema does not match')
```

### Collect All Possible Columns in a List

```python
allColumns = df1.columns + df3.columns
uniqueColumns = list(set(allColumns))
print(uniqueColumns)
```

### Add Missing Columns

```python
from pyspark.sql.functions import lit

for col in uniqueColumns:
  if col not in df1.columns:
    df1 = df1.withColumn(col, lit(None))
  if col not in df3.columns:
    df3 = df3.withColumn(col, lit(None))

display(df1)
display(df3)
```

```python
def addMissingColumns(df1, df2):
    from pyspark.sql.functions import lit
    allColumns = df1.columns + df2.columns
    uniqueColumns = list(set(allColumns))
    for col in uniqueColumns:
        if col not in df1.columns:
            df1 = df1.withColumn(col, lit(None))  
        if col not in df2.columns:
            df2 = df2.withColumn(col, lit(None))
    return df1, df2
```

```python
df1, df3 = addMissingColumns(df1, df3)
display(df1)
display(df3)
```
