### Create Sample Dataframe

```python
# sample sales data
data1 = [
    ('Product A', 100),
    ('Product B', 50),
    ('Product C', 200),
    ('Product C', 200),
    ('Product D', 150)
]

data2 = [
    ('Product A', 100),
    ('Product B', 30),
    ('Product C', 200),
    ('Product E', 80)
]

# create dataframes from the sample data
sourceDF = spark.createDataFrame(data1, ['Product', 'Quantity'])
targetDF = spark.createDataFrame(data2, ['Product', 'Quantity'])
```

### Verify Data

```python
sourceDF.display()
targetDF.display()
```

### Subtract - Not Preserving Duplicates

```python
resultDF = sourceDF.subtract(targetDF)
resultDF.display()
```

### ExceptAll - Preserves Duplicates

```python
resultDF = sourceDF.exceptAll(targetDF)
resultDF.display()
```
