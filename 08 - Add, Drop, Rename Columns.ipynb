### Add, Drop and Rename Columns in Dataframe

**Add Column**
- df.withColumn("new_column_name", value)

**Drop Column**
- df.drop("column_name")

Rename Column
- df.withColumnRenamed("old_name", "new_name")

### create Dataframe

```python
employee_Data = [(10, "Raj", "Kumar", "1999", "100", "M", 2000),
                 (20, "Rahul", "Ranjan", "2002", "200", "F", 8000),
                 (30, "Raghav", "Manish", "2010", "100", None, 6000),
                 (40, "Raja", "Singh", "2004", "100", "F", 7000),
                 (50, "Rama", "Krish", "2008", "400", "M", 1000),
                 (60, "Rasul", "Kutty", "2014", "500", "M", 5000),
                 (70, "Kumar", "Chand", "2004", "600", "M", 5000)
                 ]

employee_Schema = ["Employee_ID", "First_Name", "Last_Name", "DOJ", "Department_ID","Gender", "Salary"]

employee_DF = spark.createDataFrame(data = employee_Data, schema = employee_Schema)
display(employee_DF)
```

### Add New Column using Constant Literal

```python
from pyspark.sql.functions import lit
employee_DF_constColumn = employee_DF.withColumn("Location", lit("Mumbai")).show()
```

### Add New Column by Calculation

```python
from pyspark.sql.functions import concat

# add multiple new column
employee_DF_calcColumn = employee_DF.withColumn("Bonus", employee_DF.Salary * 0.1).withColumn("Full_Name", concat("First_Name", lit(" "), "Last_Name"))
display(employee_DF_calcColumn)
```

### Rename a Column

```python
employee_DF_renameColumn = employee_DF_calcColumn.withColumnRenamed("Full_Name", "Employee_Name").withColumnRenamed("DOJ", "Date_Of_Join")
display(employee_DF_renameColumn)
```

### Drop a Column

```python
employee_DF_Drop = employee_DF_renameColumn.drop("Employee_Name").drop("Bonus").show()
```
