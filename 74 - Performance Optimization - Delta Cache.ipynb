### Create Sample Dataframe

```python
from pyspark.sql.types import *

csvSchema = StructType([
  StructField("Year", IntegerType()),
  StructField("First_Name", StringType()),
  StructField("County", StringType()),
  StructField("Gender", StringType()),
  StructField("Count", IntegerType())
])

df = spark.read.option("sep", ",").option("header", "true").schema(csvSchema).csv("/Volumes/workspace/default/mandy/baby_names/Baby_Names_2007_2009.csv")
df.display()
```

### Create Delta Table

```python
df.write.format("delta").saveAsTable("names")
```

### Enable Delta Cahce

```python
spark.conf.set("spark.databricks.io.cache.enabled", "true")
```

```python
spark.conf.get("spark.databricks.io.cache.enabled")
```

### Select comes from DBFS

```python
%sql
select * from names
```

### Preload and Cache Entire Delta Table inot Local Disk

```python
%sql
CACHE SELECT * from names
```

```python
%sql
select * from names where Year = 2007
```
