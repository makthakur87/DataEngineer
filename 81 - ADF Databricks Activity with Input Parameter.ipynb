### JDBC Connection

```python
jdbcHostname = dbutils.secrets.get(scope="hub_vault", key="jdbcHostname")
jdbcPort = dbutils.secrets.get(scope="hub_vault", key="jdbcPort")
jdbcDatabase = dbutils.secrets.get(scope="hub_vault", key="jdbcDatabase")
jdbcUsername = dbutils.secrets.get(scope="hub_vault", key="jdbcUsername")
jdbcPassword = dbutils.secrets.get(scope="hub_vault", key="jdbcPassword")
jdbcDriver = "com.microsoft.sqlserver.jdbc.SQLServerDriver"
jdbcUrl = "jdbc:sqlserver://{jdbcHostname}:{jdbcPort}; database={jdbcDatabase}; user={jdbcUsername}; password={jdbcPassword}"
```

### Read from Azure SQL Database

```python
empDF = spark.read.format("jdbc").option("url", jdbcUrl).option("dbtable", 'dbo.emp').load()
display(empDF)
```
