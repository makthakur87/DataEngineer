### **ETL Process**
**Extract**

Step 1 - Read Tables from Azure SQL

Step 2 - JDBC Connector is used

Step 3 - Dataframes are created

**Transform**

Step 1 - Business Transformations are applied

Step 2 - Null are replaced with default values

Step 3 - Duplicates are dropped

Step 4 - Join and Aggregate are performed

**Load**

Step 1 - Mount point is created for ADLS integration

Step 2 - Data is loaded in Parquet file fomrat to target location
