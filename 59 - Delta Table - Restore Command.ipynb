### Create Delta Table

```python
from delta.tables import DeltaTable

DeltaTable.create(spark) \
    .tableName("scd2Demo") \
    .addColumn("pk1", "integer") \
    .addColumn("pk2", "string") \
    .addColumn("dim1", "integer") \
    .addColumn("dim2", "integer") \
    .addColumn("dim3", "integer") \
    .addColumn("dim4", "integer") \
    .addColumn("active_status", "string") \
    .addColumn("start_date", "timestamp") \
    .addColumn("end_date", "timestamp") \
    .execute()
```

```python
%sql
select * from scd2demo
```

### Create Delta Table Instance

```python
from delta import *

targetTable = DeltaTable.forName(spark, "scd2Demo")
```

### List Down Version History

```python
display(targetTable.history())
```

### Restore using Version Number

```python
targetTable.restoreToVersion(10)
```

```python
%sql
select * from scd2demo
```

### Restore using Timestamp

```python
targetTable.restoreToTimestamp("2025-12-21T11:47:18.000+00:00")
```

```python
%sql
select * from scd2demo
```
