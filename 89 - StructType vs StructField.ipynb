### Create Sample Dataframe

```python
structureData = [
  ('James', 111, 'HR'),
  ('Michael', 222, 'IT'),
  ('Robert', 333, 'Sales'),
  ('Maria', 444, 'IT'),
  ('Jen', 555, 'HR')
]
```

### Define Structure using Struct Type and Struct Field

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

structureSchema = StructType([
    StructField('Name', StringType(), True),
    StructField('ID', IntegerType(), True),
    StructField('Department', StringType(), True)
])
```

### Create Dataframe based on Structure Definition

```python
df = spark.createDataFrame(data = structureData, schema = structureSchema)
df.printSchema()

display(df)
```

### Nested Data

```python
structureData = [
    (('James', 'Will', 'Smith'), 111, 'HR'),
    (('Michael', 'Rose', 'Dan'), 222, 'Sales'),
    (('Robert', 'Ray', 'Williams'), 333, 'IT'),
    (('Maria', 'Anne', 'Jones'), 444, 'IT'),
    (('Jen', 'Mary', 'Brown'), 555, 'HR')
]
```

### Defind Nested Structure

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

structureSchema = StructType([
    StructField('Name', StructType([
        StructField('FirstName', StringType(), False),
        StructField('MiddleName', StringType(), True),
        StructField('LastName', StringType(), True)
    ])),
    StructField('ID', IntegerType(), True),
    StructField('Department', StringType(), True)
])
```

### Create Dataframe using Nested Structure

```python
dfNested = spark.createDataFrame(data = structureData, schema = structureSchema)
dfNested.printSchema()

display(dfNested)
```
