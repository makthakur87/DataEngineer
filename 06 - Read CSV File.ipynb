### Read CSV File
**Syntax:**

df = spark.read.format(file_type)
                .option("inferSchema", )
                .option("header", )
                .option("sep", )
                .schema(schema_df)
                .load(file_location)

**File Type:**
- CSV
- Parquet
- ORC
- JSON
- AVRO

**options:**
- header = True
- inferSchema = True
- sep = ,

**Path:**
- load(path1)
- load(path1, path2)
- load(folder)

**Schema:**
- schema(schema)

### Create Own Schema

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

schema_defind = StructType([StructField("Year", IntegerType(), True),
                            StructField("Name", StringType(), True),
                            StructField("County", StringType(), True),
                            StructField("Sex", StringType(), True),
                            StructField("Count", IntegerType(), True)
                            ])
```

```python
df = spark.read.format("csv").schema(schema_defind).option("header", "true").load("/Volumes/workspace/default/mandy/baby_names")
display(df)
```

### Alternate Schema

```python
schema_alternate = 'Year INTEGER, Name STRING, County STRING, Sex STRING, Count INTEGER'
```

```python
df = spark.read.format("csv").schema(schema_alternate).option("header", True).load("/Volumes/workspace/default/mandy/baby_names")
display(df)
```
