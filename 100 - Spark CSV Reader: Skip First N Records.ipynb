### Read CSV File and Display Data and Record Count

```python
df = (
    spark.read.format("csv")
    .option('inferSchema', True)
    .option('header', True)
    .load('/Volumes/workspace/default/mandy/baby_names/Baby_Names_2007_2009.csv')
    )

display(df)
print(df.count())
```

### Skip First 4 Records While Reading CSV File and Display Data and Record Count

```python
skipDF = (
    spark.read.format('csv')
    .option('inferSchema', True)
    .option('header', True)
    .option('sep', ',')
    .option('skipRows', 7)
    .load('/Volumes/workspace/default/mandy/baby_names/Baby_Names_2007_2009.csv')
    )

display(skipDF)
print(skipDF.count())
```
