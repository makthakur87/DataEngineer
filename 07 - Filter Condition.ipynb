### Filter
To limit the data result or extract a subset of master data
- filter(df.column != 50)
- filter(f.col('column1') > 50) & (f.col('column2') > 50)
- filter(f.col('column1') > 50) | (f.col('column2') > 50)
- filter(df.column.startswith(''))
- filter(df.column.endswith(''))
- filter(df.column.contains(''))
- filter(df.column.isNull())
- filter(df.column.isNotNull())
- filter(df.name.isin())
- filter(df.column.like('%%'))

### Create Dataframe

```python
employee_Data = [(10, "Raj Kumar", "1999", "100", "M", 2000),
                 (20, "Rahul Rajan", "2002", "200", "F", 8000),
                  (30, "Raghav", "2010", "100", None, 6000),
                   (40, "Raja Singh", "2004", "100", "F", 7000),
                    (50, "Rama Krish", "2008", "400", "M", 1000),
                     (60, "Rahul", "2014", "500", "M", 5000),
                      (70, "Kumar Chand", "2004", "600", "M", 5000)
                      ]

employee_schema = ["Employee_ID", "Name", "DOJ", "Employee_Dept_ID", "Gender", "Salary"]

employeeDF = spark.createDataFrame(data = employee_Data, schema = employee_schema)
display(employeeDF)
```

```python
display(employeeDF)

display(employeeDF.filter(employeeDF.Salary == 5000)) # ==, !=, <=, >=, <, >
```

```python
display(employeeDF)

display(employeeDF.filter(employeeDF.Salary != 5000)) # ==, !=, <=, >=, <, >
```

```python
display(employeeDF)

display(employeeDF.filter(employeeDF.Salary <= 5000)) # ==, !=, <=, >=, <, >
```

```python
display(employeeDF)

display(employeeDF.filter(employeeDF.Salary >= 5000)) # ==, !=, <=, >=, <, >
```

```python
display(employeeDF)

display(employeeDF.filter(employeeDF.Salary < 5000)) # ==, !=, <=, >=, <, >
```

```python
display(employeeDF)

display(employeeDF.filter(employeeDF.Salary > 5000)) # ==, !=, <=, >=, <, >
```

```python
display(employeeDF)

display(employeeDF.filter((employeeDF.Gender == 'F') & (employeeDF.DOJ == 2004))) # |, &
```

```python
display(employeeDF)

display(employeeDF.filter((employeeDF.Gender == 'F') | (employeeDF.DOJ == 2004))) # |, &
```

```python
display(employeeDF)

display(employeeDF.filter(employeeDF.Name.startswith('Raj'))) # startwith(), endswith(), contains()
```

```python
display(employeeDF)

display(employeeDF.filter(employeeDF.Name.endswith('h'))) # startwith(), endswith(), contains()
```

```python
display(employeeDF)

display(employeeDF.filter(employeeDF.Name.contains('Kumar'))) # startwith(), endswith(), contains()
```

```python
display(employeeDF)

display(employeeDF.filter(employeeDF.Gender.isNull())) # isNull(), isNotNull()
```

```python
display(employeeDF)

display(employeeDF.filter(employeeDF.Gender.isNotNull())) # isNull(), isNotNull()
```

```python
display(employeeDF)

display(employeeDF.filter(employeeDF.Salary.isin(1000, 5000, 8000)))  # isin(), ~
```

```python
display(employeeDF)

display(employeeDF.filter(~employeeDF.Salary.isin(1000, 5000, 8000)))  # isin(), ~
```

```python
display(employeeDF)

display(employeeDF.filter(employeeDF.Name.like("%Kumar%")))  # like(%%)
```

```python
display(employeeDF)

display(employeeDF.filter(employeeDF.Name.like("%Kumar")))  # like(%%)
```

```python
display(employeeDF)

display(employeeDF.filter(employeeDF.Name.like("Kumar%")))  # like(%%)
```
