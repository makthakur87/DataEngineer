### Create Sample Dataframe

```python
employee_data = [
    (10, "Michael Robinson", "1999-06-01", "100", 2000),
    (20, "James Wood", "2003-03-01", "200", 8000),
    (30, "Chris Andrew", "2005-04-01", "100", 6000),
    (40, "Mark Bond", "2008-10-01", "100", 7000),
    (50, "Steve Watson", "1996-02-01", "400", 1000),
    (60, "Mathews Simon", "1998-11-01", "500", 5000),
    (70, "Peter Paul", "2011-04-01", "600", 5000)
]

employee_schema = ["Employee_Id", "Employee_Name", "DOJ", "Department_Id", "Salary"]

empDF = spark.createDataFrame(employee_data, employee_schema)
display(empDF)
```

### First Method of Split

```python
from pyspark.sql.functions import split

df1 = empDF.withColumn("First_Name", split("Employee_Name", " ").getItem(0)) \
      .withColumn("Last_Name", split("Employee_Name", " ").getItem(1))
      
display(df1)

```

### Second Split Method

```python
import pyspark

split_col = pyspark.sql.functions.split(empDF["Employee_Name"], " ")

df2 = empDF.withColumn('First_Name', split_col.getItem(0)) \
        .withColumn('Last_Name', split_col.getItem(1))

display(df2)
```

### Third Split Method

```python
split_col = pyspark.sql.functions.split(empDF["DOJ"], "-")
df3 = empDF.select("Employee_Id", "Employee_Name", "Department_Id", "Salary", 
                   split_col.getItem(0).alias("Joining_Year"), split_col.getItem(1).alias("Joining_Month"), 
                   split_col.getItem(2).alias('Joining_Day'))
display(df3)
```

### Combine Multiple Split

```python
df4 = empDF.withColumn('First_Name', split(empDF['Employee_Name'], ' ').getItem(0)) \
        .withColumn('Last_Name', split(empDF['Employee_Name'], ' ').getItem(1)) \
            .withColumn('Joining_Year', split(empDF['DOJ'], '-').getItem(0)) \
                .withColumn('Joining_Month', split(empDF['DOJ'], '-').getItem(1)) \
                    .withColumn('Joining_Day', split(empDF['DOJ'], '-').getItem(2))
display(df4)
```

### Split and Drop Splitted Columns

```python
df5 = empDF.withColumn('First_Name', split(empDF['Employee_Name'], ' ').getItem(0)) \
        .withColumn('Last_Name', split(empDF['Employee_Name'], ' ').getItem(1)) \
            .withColumn('Joining_Year', split(empDF['DOJ'], '-').getItem(0)) \
                .withColumn('Joining_Month', split(empDF['DOJ'], '-').getItem(1)) \
                    .withColumn('Joining_Day', split(empDF['DOJ'], '-').getItem(2)) \
                        .drop(empDF['Employee_Name']) \
                            .drop(empDF['DOJ'])
display(df5)
```
