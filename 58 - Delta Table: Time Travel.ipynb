### Pyspark Approaches

### Method 1 - Pyspark - Timestamp + Table

```python
df = spark.read \
    .format("delta") \
    .option("timestampAsOf", "2025-12-21T12:01:15.000+00:00") \
    .table("scd2demo")

display(df)
```

### Method 2 - Pyspark - Timestamp + Path

```python
df = spark.read \
    .format("delta") \
    .option("timestampAsOf", "2025-12-21T12:01:15.000+00:00") \
    .load(".workspace.default.scd2demo")

display(df)
```

### Method 3 - Pyspark - Version + Path

```python
df = spark.read \
    .format("delta") \
    .option("versionAsOf", 5) \
    .load(".workspace.default.scd2demo")

display(df)
```

### Method 4 - Pyspark - Version + Table

```python
df = spark.read \
    .format("delta") \
    .option("versionAsOf", 8) \
    .table("scd2demo")

display(df)
```
