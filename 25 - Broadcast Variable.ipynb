### Create Sample Dataframe

```python
Transaction = [
  (100, 'Cosmetic', 150),
  (200, 'Apparel', 250),
  (300, 'Shirt', 400),
  (400, 'Trouser', 500),
  (500, 'Socks', 20),
  (100, 'Belt', 70),
  (200, 'Cosmetic', 250),
  (300, 'Shoe', 400),
  (400, 'Socks', 25),
  (500, 'Shorts', 100),
  ]

transactionDF = spark.createDataFrame(data = Transaction, schema = ['Store_ID', 'Item', 'Amount'])
transactionDF.show()
```

### Create Sample Dimension Table

```python
Stores = [
  (100, 'Store_London'),
  (200, 'Store_Paris'),
  (300, 'Store_Frankfurt'),
  (400, 'Store_Stockholm'),
  (500, 'Store_Oslo'),
]
storeDF = spark.createDataFrame(data = Stores, schema = ['Store_ID', 'Store_Name'])
storeDF.show()
```

```python
from pyspark.sql.functions import broadcast
joinDF = transactionDF.join(broadcast(storeDF), transactionDF['Store_ID'] == storeDF['Store_ID'])

joinDF.show()
```
