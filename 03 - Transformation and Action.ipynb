### Transformation
Transformation is kind of operation which will transform dataframe from one form to another. We will get new dataframe after execution of each transformation operation. Transformation is lazy evaluation based on DAG. e.g. Filter, Union etc.

### Action
When we want to work with actual data, call the action. Action returns the data to driver for display or storage into storage layer. e.g. count, collect, save etc.

### Read the Source Data to create Dataframe

```python
df = spark.read.format("csv").option("inferSchema", True).option("header", True).option("sep", ",").load("/Volumes/workspace/default/mandy/baby_names")
display(df)
```

### Peform Narrow Transformation1 : Filter

```python
df1 = df.filter(df.County=="Franklin County")
```

### Peform Narrow Transformation2 : Filter

```python
df2 = df.filter(df.County == "Lake County")
```

### Peform Narrow Transformation3 : Filter

```python
df3 = df1.union(df2)
```

### Perform Wide Transformation : GroupBy

```python
df4 = df.groupBy("Year").count()
```

### Calling Action Collect

```python
df4.collect()
```
