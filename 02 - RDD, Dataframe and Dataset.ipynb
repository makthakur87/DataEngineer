### RDD, Dataframe and Dataset
- They all are APIs provided by Spark for developers for data processing and analytics

- In terms of functionality, all are same and returns same output for provided input data. But they differ in the way of handling and processing data. So. there is difference in terms of performance, user convenienve and language support etc.

- Users can choose any of the API while working with Spark
