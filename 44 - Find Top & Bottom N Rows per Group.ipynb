### Find Top and Bottom N Rows per Group
### Create Sample Dataframe

```python
data_student = [
  ('Michael', 'Physics', 80, 'P', 90),
  ('Michael', 'Chemistry', 67, 'P', 90),
  ('Michael', 'Mathematics', 78, 'P', 90),
  ('Nancy', 'Physics', 30, 'F', 80),
  ('Nancy', 'Chemistry', 59, 'P', 80),
  ('Nancy', 'Mathematics', 75, 'P', 80),
  ('David', 'Physics', 90, 'P', 70),
  ('David', 'Chemistry', 87, 'P', 70),
  ('David', 'Mathematics', 97, 'P', 70),
  ('John', 'Physics', 33, 'F', 60),
  ('John', 'Mathematics', 28, 'F', 60),
  ('John', 'Chemistry', 52, 'P', 60),
  ('Blessy', 'Physics', 89, 'P', 75),
  ('Blessy', 'Physics', 76, 'P', 75),
  ('Blessy', 'Physics', 63, 'P', 75)
]

schema = ['Name', 'Subject', 'Mark', 'Status', 'Attendance']
df = spark.createDataFrame(data=data_student, schema=schema)
display(df)
```

### Create Rank within Each Group of Name

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number, col
widnowDept = Window.partitionBy('Name').orderBy(col('Mark').desc())
df2 = df.withColumn('Row', row_number().over(widnowDept)).orderBy('Name', 'Row')
display(df2)
```

### Get Top N Rows per Group of Name

```python
df3 = df2.filter(col('Row') <= 1)
display(df3)
```

### Create Rank within Each Group of Subject

```python
widnowDept = Window.partitionBy('Subject').orderBy(col('Mark').desc())
df4 = df.withColumn('Row', row_number().over(widnowDept)).orderBy('Name', 'Row')
display(df4)
```

### Get Top N Rows per Group of Subject

```python
df5 = df4.filter(col('Row') <= 1)
display(df5)
```

### Reverse Rank to Get Bottom N Rows per Group

```python
widnowDept = Window.partitionBy('Subject').orderBy(col('Mark'))
df6 = df.withColumn('Row', row_number().over(widnowDept)).orderBy('Name', 'Row')
display(df6)
```

### Get Bottom N Rows per Group

```python
df7 = df6.filter(col('Row') <= 1)
display(df7)
```
