### Create Sample Dataframe

```python
simpleData = (
  (111, 'James', 8),
  (222, 'Mike', 5),
  (111, 'James', 8),
  (222, 'Mike', 12),
)

columns = ['ID', 'Name', 'Score']
df = spark.createDataFrame(data = simpleData, schema = columns)

display(df)
```

### Distinct to Remove Duplicate

```python
df.distinct().display()
```

### Distinct with Subset

```python
df.select('ID', 'Name').distinct().display()
```

### Drop Duplicate without Subset

```python
df.dropDuplicates().display()
df.drop_duplicates().display()
```

### Drop Duplicate with Subset

```python
df.dropDuplicates(["ID", "Name"]).display()

df.drop_duplicates(["ID", "Name"]).display()
```
