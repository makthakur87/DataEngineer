### **Corrupt Record Handling**
**Permissive** - Include corrupt record in separate column

**Drop Malformed** - Ignore corrupt records

**Fail Fast** - Throw exception if corrupt record

![](Images/16 Corrupt Record Handling.jpg)

### Syntax

![](Images/16 Syntax.jpg)

```python
df = spark.read.format("csv").option("header", "true").option("inferschema", "false").load("/Volumes/workspace/default/mandy/Production_Data_Corrupt.csv")
display(df)
```

### **Define Schema**

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType
schema = StructType([\
    StructField("Month", StringType(), True), \
    StructField("Emp_Count", IntegerType(), True), \
    StructField("Production_Unit", IntegerType(), True), \
    StructField("Expense", IntegerType(), True), \
    StructField("_corrupt_record", StringType(), True)
    ])
```

### **Permissive Mode**

```python
df = spark.read.format("csv").option("mode", "PERMISSIVE").option("header", "true").schema(schema).load("/Volumes/workspace/default/mandy/Production_Data_Corrupt.csv")
display(df)
```

### **DropMalformed Mode**

```python
df = spark.read.format("csv").option("mode", "DROPMALFORMED").option("header", "true").schema(schema).load("/Volumes/workspace/default/mandy/Production_Data_Corrupt.csv")
display(df)
```

### **FailFast Mode**

```python

df = spark.read.format("csv").option("mode", "FAILFAST").option("header", "true").schema(schema).load("/Volumes/workspace/default/mandy/Production_Data_Corrupt.csv")
display(df)
```
