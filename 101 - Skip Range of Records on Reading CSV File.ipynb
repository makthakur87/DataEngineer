### Write a Program to Skip Records on Specific Ranges in the Middle of the CSV File, For Example Rows 11 to 20

### Verify Data in CSV File

```python
%fs
head /Volumes/workspace/default/mandy/baby_names/Baby_Names_2007_2009.csv
```

### Step 1 - Create Dataframe by Reading All Rows

```python
fullDF = (
    spark.read.format("csv")
    .option('inferSchema', True)
    .option('header', True)
    .load('/Volumes/workspace/default/mandy/baby_names/Baby_Names_2007_2009.csv')
    )

display(fullDF)
print(fullDF.count())
```

### Step 2 - Create Dataframe by Reading Only from Starting Row of Specific Range

```python
skipStartDF = (
    spark.read.format('csv')
    .option('inferSchema', True)
    .option('header', True)
    .option('sep', ',')
    .option('skipRows', 10)
    .load('/Volumes/workspace/default/mandy/baby_names/Baby_Names_2007_2009.csv')
    )

display(skipStartDF)
print(skipStartDF.count())
```

### Step 3 - Create Dataframe by Reading Data only from Ending Row of Specific Range

```python
skipEndDF = (
    spark.read.format('csv')
    .option('inferSchema', True)
    .option('header', True)
    .option('sep', ',')
    .option('skipRows', 20)
    .load('/Volumes/workspace/default/mandy/baby_names/Baby_Names_2007_2009.csv')
    )

display(skipEndDF)
print(skipEndDF.count())
```

### Step 4 - Find the Delta between fullDF and SkipStartDF

```python
deltaDF = fullDF.subtract(skipStartDF)
display(deltaDF)
```

### Step 5 - Combine deltaDF and skipEndDF

```python
finalDF = deltaDF.union(skipEndDF)
display(finalDF.orderBy('Year'))
```

### UDF Combining All 5 Steps

```python
def skipRowsRangeCSV(filePath, startPos, endPos):
    fullDF = (
        spark.read.format('csv')
        .option('inferSchema', True)
        .option('header', True)
        .option('sep', ',')
        .load(filePath)
    )

    skipStartDF = (
        spark.read.format('csv')
        .option('inferSchema', True) 
        .option('header', True)
        .option('sep', ',')
        .option('skipRows', startPos)
        .load(filePath)
    )

    skipEndDF = (
        spark.read.format('csv')
        .option('inferSchema', True) 
        .option('header', True)
        .option('sep', ',')
        .option('skipRows', endPos)
        .load(filePath)
    )

    deltaDF = fullDF.subtract(skipStartDF)
    finalDF = deltaDF.union(skipEndDF)

    return finalDF
```

```python
outputDF = skipRowsRangeCSV(filePath='/Volumes/workspace/default/mandy/baby_names/Baby_Names_2007_2009.csv', startPos=10, endPos=20)
display(outputDF.orderBy('Year'))
```
