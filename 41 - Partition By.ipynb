### Create Sample Dataframe

```python
df = spark.read.format("csv").option("header", "true").load("/Volumes/workspace/default/mandy/baby_names/")
display(df)
```

### Distinct Year List and Count for Each Year

```python
df.groupBy("Year").count().show(truncate=False)
```

### Partition by One Key Column

```python
df.write.option("header", "true") \
    .partitionBy('Year') \
    .mode("overwrite") \
    .csv("/Volumes/workspace/default/mandy/baby_names_output/")
```

### Partition by Multiple Key Columns

```python
# dbutils.fs.rm("/Volumes/workspace/default/mandy/baby_names_output/", True)
dbutils.fs.mkdirs("/Volumes/workspace/default/mandy/baby_names_output")
```

```python
df.write.option("header", "true") \
    .partitionBy('Year', 'Sex') \
    .mode("overwrite") \
    .csv("/Volumes/workspace/default/mandy/baby_names_output/")
```

### Partition by Key Column along with Maximum Number of Records foe Each Partition

```python
# dbutils.fs.rm("/Volumes/workspace/default/mandy/baby_names_output/", True)
dbutils.fs.mkdirs("/Volumes/workspace/default/mandy/baby_names_output")
```

```python
df.write.option("header", True) \
    .option('maxRecordsPerFile', 300) \
    .partitionBy('Year') \
    .mode("overwrite") \
    .csv("/Volumes/workspace/default/mandy/baby_names_output/")
```
