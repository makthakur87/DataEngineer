**What is Explode**
It is a pyspark function that returns a new row for each element in the given array or map. uses the default column name col for elements in the array and key and value for elements in the map

**Variants**

**Explode** -  When an array is passed to this function, it created a new row for each element in the array. When a map is passed, it creates two new columns one for key and one for value and each element in map split into  the rows. if the array or map is null, that row is eliminated.

**Explode_outer** - unlike explode, if the array or map is null, explode_outer returns null.

**Posexplode** -  When an array or map is passed, creates positional column also for each element. Ignores the null elements

**Posexplode_outer** - Unlike posexplode, if the array or map is null, explode_outer returns null

##  **Create Dataframe with Array Column**

```python
array_appliance = [
    ('Raja', ['TV', 'Refrigerator', 'Oven', 'AC']),
    ('Raghav', ['AC', 'Washing Machine', None]),
    ('Ram', ['Grinder', 'TV']),
    ('Ramesh', ['TV', 'Refrigerator', None]),
    ('Rajesh', None)
    ]

df_app = spark.createDataFrame(data = array_appliance, schema = ['Name', 'Appliances'])
df_app.printSchema()
display(df_app)
```

## **Create Dataframe with Map Column**

```python
map_appliance = [
    ('Raja', {'TV': 'LG', 'Refrigerator': 'Samsung', 'Oven': 'Phillips', 'AC': 'Voltas'}),
    ('Raghav',  {'AC': 'Samsung', 'Washing Machine': 'LG'}),
    ('Ram', {'Grinder': 'Preethi', 'TV': ''}),
    ('Ramesh', {'Refrigerator': 'LG', 'TV': 'Croma'}),
    ('Rajesh', None)
    ]

df_map = spark.createDataFrame(data = map_appliance, schema = ['Name', 'Brand'])
df_map.printSchema()
display(df_map)
```

## **Explode Array Field**

```python
from pyspark.sql.functions import explode
df2 = df_app.select(df_app.Name, explode(df_app.Appliances))

df_app.printSchema()
display(df_app)

df2.printSchema()
display(df2)
```

## **Explode Map Field**

```python
from pyspark.sql.functions import explode_outer
df3 = df_map.select(df_map.Name, explode(df_map.Brand))

df_map.printSchema()
display(df_map)

df3.printSchema()
display(df3)
```

## **Display Outer to Consider Null Values**

```python
from pyspark.sql.functions import explode_outer
display(df_app.select(df_app.Name, explode_outer(df_app.Appliances)))

display(df_map.select(df_map.Name, explode_outer(df_map.Brand)))
```

## **Positional Explode**

```python
from pyspark.sql.functions import posexplode
display(df_app.select(df_app.Name, posexplode(df_app.Appliances)))

display(df_map.select(df_map.Name, posexplode(df_map.Brand)))
```

## **Positinal Explode with Null**

```python
from pyspark.sql.functions import posexplode_outer
display(df_app.select(df_app.Name, posexplode_outer(df_app.Appliances)))

display(df_map.select(df_map.Name, posexplode_outer(df_map.Brand))) 
```
