### Create Delta Table

```python
from delta.tables import *

DeltaTable.create(spark) \
  .tableName('employee_demo_new') \
  .addColumn('emp_id', 'INT') \
  .addColumn('emp_name', 'STRING') \
  .addColumn('gender', 'STRING') \
  .addColumn('salary', 'INT') \
  .addColumn('dept', 'STRING') \
  .property('description', 'table created for demo purpose') \
  .execute()
```

```python
%sql
select * from employee_demo_new
```

### Poplulate Sample Data

```python
%sql
insert into employee_demo_new values(100, 'Stephen', 'M', 2000, 'IT');
insert into employee_demo_new values(200, 'Phillip', 'M', 8000, 'HR');
insert into employee_demo_new values(300, 'Lara', 'F', 6000, 'SALES');
insert into employee_demo_new values(400, 'Nike', 'M', 4000, 'IT');
insert into employee_demo_new values(500, 'Sarah', 'F', 9000, 'HR');
insert into employee_demo_new values(600, 'Serena', 'F', 5000, 'SALES');
insert into employee_demo_new values(700, 'Mark', 'M', 7000, 'SALES');
```

```python
%sql
select * from employee_demo_new
```

### Method 1 - Pyspark Delta Table Instance

```python
from delta.tables import *
from pyspark.sql.functions import *
deltaTable = DeltaTable.forName(spark, "employee_demo_new")

# Declare the predicate by using a SQL formatted String
deltaTable.delete("emp_id = 200")
```

```python
%sql
select * from employee_demo_new
```

### Multiple Conditions using SQL Predicate

```python
deltaTable.delete("emp_id = 400 and gender = 'M'")
```

```python
%sql
select * from employee_demo_new
```

### Method 1 - Pyspark Delta Table Instance - Spark SQL Predicate

```python
deltaTable.delete(col("emp_id") == 500)
```
