### **Union**
To combine two dataframes. The schema of both dataframes should match. Removes the duplocate records from resultant dataframe until spark version 2.0.0. So, duplicate can be removed manually by dropDuplicates()

### **Syntax:**
DF_Union = DF1.union(DF2)

### **UnionAll**
Same as union but retains duplicate records as well in resultant dataframe.

### **Syntax:**
DF_UnionAll = DF1.unionAll(DF2)

### **Spark Version**

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
print(spark.version)
```

### **Create Dataframe DF1**


```python
employee_data = [(100, "Stephen", "1999", "100", "M", 2000),
                 (200, "Phillip", "2002", "200", "M", 8000),
                 (300, "John", "2010", "100", "", 6000)]
employee_schema = ["Employee_ID", "Name", "DOJ", "Employee_Dept_ID", "Gender", "Salary"]

DF1 = spark.createDataFrame(data=employee_data, schema=employee_schema)
display(DF1)
```

### **Create Dataframe DF2**

```python
employee_data = [(300, "John", "2010", "100", "", 6000),
                 (400, "Nancy", "2008", "400", "F", 1000),
                 (500, "Rosy", "2014", "500", "F", 5000)]
employee_schema = ["Employee_ID", "Name", "DOJ", "Employee_Dept_ID", "Gender", "Salary"]

DF2 = spark.createDataFrame(data=employee_data, schema=employee_schema)
display(DF2)
```

### **Union**

```python
df_union = DF1.union(DF2)
display(df_union)
```

### **Drop Duplicates**

```python
display(df_union.dropDuplicates())
```

### **UnionAll**

```python
df_unionAll = DF1.unionAll(DF2)
display(df_unionAll)
# display(df_unionAll.dropDuplicates())
```

### **Schema Mismatch**

```python
DF3 = DF2.select(DF2.Employee_ID, DF2.Name, DF2.DOJ, DF2.Employee_Dept_ID, DF2.Gender)
display(DF3)
```

```python
DF_Schema_Mismatch = DF1.union(DF3)
display(DF_Schema_Mismatch)
```
