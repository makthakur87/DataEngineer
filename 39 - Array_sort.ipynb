### Create Sample Dataframe

```python
empDF = [
  ('John', [4, 6, 7, 9, 2]),
  ('David', [7, 5, 1, 4, 7, 1]),
  ('Mike', [3, 9, 1, 6, 2])
]

df = spark.createDataFrame(data = empDF, schema = ['Name', 'Array_1'])
df.show()
```

### Applry Array_sort

```python
from pyspark.sql import functions as F
df_sort = df.withColumn("Sorted", F.array_sort(df["Array_1"]))
df_sort.show()
```
